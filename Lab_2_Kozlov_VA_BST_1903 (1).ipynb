```python
import numpy
def binarniy(mass, n):
    rigth=int(len(mass))
    rigth=rigth-1
    left=int(0)
    while (n > mass[int(left+((rigth-left)/2))]) or (n < mass[int(left+((rigth-left)/2))]):
       # print(int((rigth-left)/2))
        if n>mass[int(left+((rigth-left)/2))]:
            left=int(left+((rigth-left)/2))
        if n<mass[int(left+((rigth-left)/2))]:
            rigth=int(left+((rigth-left)/2))
    if mass[int(left+((rigth-left)/2))]==n:
        return int(left+((rigth-left)/2))
    return -1
```

```python
class TreeNode:
    def __init__(self, value):
        self.value=value
        self.right=None
        self.left=None
    def addTo(self, mas):
        if mas<self.value:
            if self.left is None:
                self.left=TreeNode(mas)
            else:
                self.left.addTo( mas)
        else:
            if self.right is None:
                self.right=TreeNode(mas)
            else:
                self.right.addTo(mas)
        
            
    def pre_order(self, nl):
        if nl and self and self.value is not None:
            if nl>self.value and self.right:
                self=self.right.pre_order(nl)
            elif nl<self.value and self.left:
                self=self.left.pre_order(nl)
            elif(nl==self.value):
                print("Элемент найден")
                return self
            else:
                print("Элемент не найдет")
                return None
            
        else:
            print("Элемент не найдет")
            return None
        return self
    
    def deleted(self):
        if self.right is not None:
            self.value=self.right.value
            self.right.deleted()
        elif self.left is not None:
            self.value=self.left.value
            self.left.deleted()
        elif (self.left is None) and (self.right is None):
            self.value=None
    
    def delit(self, i):
        self.pre_order(i).deleted()
        
    def pr(self):
        print(self.value)
        if self.right:
            self.right.pr()
        if self.left:
            self.left.pr()
            
    mass=[3, 6, 8, 7, 9]
    tree=TreeNode(mass[0])
    for i in range(1, len(mass)):
        tree.addTo(mass[i])
    tree.pr()
    print("search", tree.pre_order(7))
    tree.delit(3)
    tree.pr()
    print(tree.pre_order(3))
```

```python
def search_Fibonacci(mass, n):
    fib1=0
    fib2=1
    fibm=fib1+fib2
    while len(mass)>fibm:
        fib1=fib2
        fib2=fibm
        fibm=fib1+fib2
    index=-1
    while (fibm>1):
        i=min(index+fib1, len(mass)-1)
        if (mass[i]<n):# если n больше элемента, то переместить три переменные Фибоначчи на одну вниз и сбросить индекс до i
            fibm=fib2
            fib2=fib1
            fib1=fibm-fib2
            index=i
        elif (mass[i]>n):#если n меньше элемента, то переместить три переменные Фибоначчи на 2 вниз
            fibm=fib1
            fib2=fib2-fib1
            fib1=fibm-fib2
        else:#
            return i
    if (fib2 and index<(len(mass)-1) and mass[index+1]==n):
        return index+1
    return -1
```

```python
def Interpolation_search(mass, n):
    rigth=int(len(mass)-1)
    left=0
    while rigth>=left and n>=mass[left] and n<=mass[rigth]:
        index=left+int(((float(rigth-left)/(mass[rigth]-mass[left]))*(n-mass[left])))
        if (mass[index]==n):
            return index
        elif (mass[index]<n):
            left=index+1
        else:
            rigth=index-1
    return -1
```

```python
import random
#Простое рехэширование
class prost_rehash:
    # Конструктор, создание словаря
    def __init__(self):
        self.rhash=[None]*256
        
    def keys(self, element): # Формирование ключа
        key=0
        for i in range(len(element)):
            key=key+ord(element[i])
        return int(key%256)
    
    def add(self, element): # Добавление элемента 
        key=self.keys(element)
        while self.rhash[key] is not None:
            key=key+1
        self.rhash[key]=element
                   
    def search(self, element): # Поиск элемента
        key=self.keys(element)
        while self.rhash[key] is not None:
            if self.rhash[key]==element:
                return key
            else:
                key=key+1
        return None
    
    def deleted(self, element): # Удаление элемента
        key=self.search(element)
        while key is not None and self.rhash[key] is not None:
            if self.rhash[key]==element:
                del self.rhash[key]
                key=int(key+1)
                while key<len(self.rhash) and self.rhash[key] is not None:
                    el=self.rhash.pop(key)
                    self.add(el)
                    key=key+1
                return 1
            else:
                key=key+1
        return -1
    
    def pr(self): #
        for  key, i in enumerate(self.rhash):
            if self.rhash[key] is not None:
                print(key, " ", i)
#
```

```python
class random_rehash():
    # Конструктор, создание словаря
    def __init__(self):
        self.rhash=[None]*256
    
    def rand(self, element): #генерация псевдослучайного числа
        key=int(0)
        for i in range(len(element)):
            key=key+ord(element[i])
        return key
    
    def keys(self, key, l): # Формирование ключа
        random.seed(l)
        return int(key+((random.random()*10000000000000000)%1000))
            
    
    def add(self, element): # Добавление элемента 
        l=int(0)
        key=self.rand(element)
        key1=self.keys(key, l)%256
        while key1<len(self.rhash) and self.rhash[key1] is not None:
            l=l+1
            key1= self.keys(key, l)%256
        if key1<len(self.rhash):
            self.rhash[key1]=element
        else:
            print("Таблица заполнена")
                   
    def search(self, element): # Поиск элемента
        l=int(0)
        key=self.rand(element)
        key1=self.keys(key, l)%256
        while key1<len(self.rhash) and self.rhash[key1] is not None:
            if self.rhash[key1]==element:
                return key1
            else:
                l=l+1
                key1=self.keys(key, l)
        return None
    def deleted(self, element): # Удаление элемента
        l=int(0)
        keyn=self.rand(element)
        key1=self.keys(keyn, l)%256
        key=self.search(element)
        if key is not None:
            while key is not key1:
                l=l+1
                key1=self.keys(keyn, l)%256
                self.rhash[key]=None
                l=l+1
                key1=self.keys(keyn, l)%256
            while key1<len(self.rhash) and self.rhash[key1] is not None:
                el=self.rhash[key1]
                self.rhash[key1]=None
                self.add(el)
                l=l+1
                key1=self.keys(keyn, l)%256
            return "Элемент удален"
        else:
            return "Элемент не найден"
                 
    def pr(self): #
        for key, i in enumerate(self.rhash):
            if self.rhash[key] is not None:
                print(key, " ", i)
```

```python
class chain_rehash:
    # Конструктор, создание словаря
    def __init__(self):
        self.rhash=[[]*0 for i in range(10)]
    
    def add(self, element):  # Добавление элемента
        key=int(0)
        for i in range(len(element)):
            key=key+ord(element[i])
        key=key%10
        self.rhash[key].append(element)
        
        
    def search(self, element): #Поиск
            key=int(0)
            for i in range(len(element)):
                key=key+ord(element[i])
            key=key%10
            if self.rhash[key] is not None:
                for i in range(len(self.rhash[key])):
                    if self.rhash[key][i]==element:
                        return key, i
            return None, None
            
    def deleted(self, element):
            key, i=self.search(element)
            if key is not None:
                del(self.rhash[key][i])
                print("Элемент успешно удален")
            else:
                print("Элемент не найден")
                return -1
    
    def pr(self): #вывод
        for key in range(len(self.rhash)):
            for i in range(len(self.rhash[key])):
                if self.rhash[key][i] is not None:
                    print("(ключ)", key, "- (Элемент)", self.rhash[key][i])
#
```

```python
n=int(input())
if n%6==2:
    even=list(range(1,n,2))
    odd=list(range(2,n,2))
    odd[1]=0
    odd.append(4)
elif n%6==3:
    even=list(range(3, n, 2))
    odd=list(range(4, n, 2))
    even.append(1)
    odd.append(0)
    odd.append(2)
else:
    even=list(range(1,n,2))
    odd=list(range(0,n,2))
rez=even+odd
row=["."]*n
for i in range(n):
    j=rez[i]
    row[j]="Ф"
    print(*row)
    row[j]="."
i=n-1
print("\n")
while i>=0:
    j=rez[i]
    row[j]="Ф"
    print(*row)
    row[j]="."
    i-=1
```
