```python
class Stack:
    def __init__(self):
        self.stek=list()

    def isEmpty(self):
        return len(self.stek)==0

    def add(self, item):
        self.stek.append(item)

    def pop(self):
        if self.isEmpty == 0:
            return None
        else:
            return self.stek.pop()

    def peek(self):
        if self.isEmpty():
            return None
        else:
            return self.stek[-1]

    def __len__(self):
        return len(self.stek)
```

# Реализация дека

```python
class Deque:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def addRigth(self, item):
        self.items.append(item)

    def addLeft(self, item):
        self.items.insert(0,item)

    def removeRigth(self):
        return self.items.pop()

    def removeLeft(self):
        return self.items.pop(0)

    def size(self):
        return len(self.items)
        
    def peekLeft(self):
        if self.isEmpty():
            return None
        else:
            return self.items[0]
        
        
    def peekRigth(self):
        if self.isEmpty():
            return None
        else:
            return self.items[-1]
```

# Задание 1

```python
import copy
import codecs
#чтение и запись в файл

# Вывод из файла в стек
def add_with_stek(file):
    st=Stack()
    try:
        f=open(file,'r')
        t=f.readlines()
        f.close()
        for i in range(len(t)):
            st.add(t[i])
        return st
    except:
        print("Error, не удалось найти файл")

        
# Вывод файла в переменную типа string
def add_with_str(file):
    rez=""
    try:
        f=codecs.open(file+'.txt','r', encoding='utf-8')
        t=f.readlines()
        f.close()
        for i in range(len(t)):
            rez=rez+t[i]
        return rez
    except:
        print("Error, не удалось найти файл")
        
# Вывод из файла в дек
def add_with_dek(file):
    st=Deque()
    try:
        f=open(file,'r')
        t=f.readlines()
        f.close()
        for i in range(len(t)):
            st.addLeft(t[i])
        return st
    except:
        print("Error, не удалось найти файл")

# Запись из дека в файл
def add_file(file, dek):
    try:
        f=open(file,"w", encoding="utf-8")
        while not dek.isEmpty():
            f.write(dek.removeLeft())
        f.close()
    except:
        print("Ошибка")
        return None

def add_file_stek(file, stek):
    try:
        f=open(file+'.txt',"w", encoding="utf-8")
        while not stek.isEmpty():
            f.write(stek.pop())
        f.close()
    except:
        print("Ошибка")
        return None

# Ввод из переменной в файл
def add_file_str(file, text):
    try:
        f=open(file+'.txt',"w", encoding="utf-8")
        f.write(text)
        f.close()
    except:
        print("Ошибка")
        return None
```

```python
import random

def shifr():
    alphabet = list('абвгдеёжзийклмнопрстуфхцчшщъыьэюя')
    random.shuffle(alphabet)
    alphabet = ''.join(alphabet)
    #print(alphabet)
    keyRing = Deque()
    for letter in alphabet:
        keyRing.addRigth(letter)
    return keyRing
```
