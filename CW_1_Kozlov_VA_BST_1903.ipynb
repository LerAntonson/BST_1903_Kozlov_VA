# Задание 1 - Треугольник с максимальным периметром

##### Массив A состоит из целых положительных чисел ­ длин отрезков. Составьте из трех отрезков такой треугольник, чтобы его периметр был максимально возможным. Если невозможно составить треугольник с положительной площадью функция возвращает 0.

```python
def vstavka(arr, length):
    for i in range(length):
        key= arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
        
def Perimetr(a, b, c):
    if (a+b)>c and (a+c)>b and (b+c)>a:
        return int(a+b+c)
    else:
        return 0
length = int(input('Введите длину массива '))
massive=[]
for i in range(length):
    massive.append(int(input()))
vstavka(massive, length)
for i in range(length-2):
    if Perimetr(massive[i], massive[i+1], massive[i+2])>0:
        print("Периметр=", Perimetr(massive[i], massive[i+1], massive[i+2]))
        break  
```

##### Дана строка s, вернуть самую длинную полиндромную подстроку в s

```python
def palindrome(x):
    if len(x) <= 1: 
        return False
    return x == x[::-1] 
def zadacha_2(s):
    last = len(s)
    lst = [] 
    for i in range(last):
        for j in range(i, last): 
            b = s[i:j+1] 
            if palindrome(b):
                lst.append(b)
    return lst[0] 
st = input("Введите строку: ")
print("Самая длинная палиндромная строка:")
print(zadacha_2(st))
```

# Задание 7 - Интервалы

##### Дан массив отрезков intervals, в котором intervals[i] = [starti, endi], некоторые отрезки могут пересекаться. Напишите функцию, которая объединяет все пересекающиеся отрезки в один и возвращает новый массив непересекающихся отрезков.

```python
def intervals(interv):
    interv.sort()
    i=0
    while i <len(interv)-1:
        if interv[i][1]>=interv[i+1][0]:
            a=interv.pop(0)
            b=interv.pop(0)
            interv.insert(0, [a[0],b[1]])
        else:
            i+=1
    return interv

intervals( [[1,4],[4,5]])
```
